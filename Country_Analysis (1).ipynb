```python
df.rename(columns={
    'GDP (in Trillions USD)': 'GDP (T USD)',
    'GDP per Capita (in USD)': 'GDP/Capita (USD)'
}, inplace=True)
```

```python
# Group by 'Country' and calculate the sum of 'GDP (T USD)' for each country
gdp_sum_by_country = df.groupby('Country')['GDP (T USD)'].sum()

# Sort the resulting Series by GDP values in ascending order
data = gdp_sum_by_country.sort_values(ascending = False)

# Display the sorted Series
print(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set(style="whitegrid")

# Sort the DataFrame by 'GDP (T USD)' in ascending order
df_sorted = df.sort_values(by='GDP (T USD)' , ascending = False)

# Create a barplot with the sorted DataFrame
plt.figure(figsize=(6, 4))
barplot = sns.barplot(x='Country', y='GDP (T USD)', data=df_sorted)

# Add labels and title
plt.xlabel('Country')
plt.ylabel('GDP (in Trillions USD)')
plt.title('GDP by Country')

# Rotate x labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# List of columns to sum (absolute values)
sum_columns = [
    'GDP (T USD)', 
    'Population (in Millions)', 
    'Number of International Visitors (in Millions)', 
    'Tourism Revenue (in Billion USD)'
]

# List of columns to mean (percentage and per-capita values)
mean_columns = [
    'GDP/Capita (USD)', 
    'Inflation Rate (%)',
    'Population Growth Rate (%)', 
    'Urban Population (%)', 
    'Life Expectancy (Years)', 
    'Healthcare Expenditure per Capita (USD)',
    'Poverty Rate (%)', 
    'Agricultural Land (%)', 
    'Unemployment Rate (%)', 
    'Labor Force Participation Rate (%)', 
    'Crime Rate (per 100,000)', 
    'Corruption Perception Index', 
    'Freedom of Press Index', 
    'Voting Participation Rate (%)'
]

# Group by 'Country' and calculate sum and mean
df_sum = df.groupby('Country')[sum_columns].sum()
df_mean = df.groupby('Country')[mean_columns].mean()

# Combine sum and mean results
df_summary = pd.concat([df_sum, df_mean], axis=1)

# Sort all columns by descending order for each column
df_summary_sorted = df_summary.sort_values(by=list(df_summary.columns), ascending=False)

# Display the resulting DataFrame
print(df_summary_sorted)
```

```python
# Filter the DataFrame for India and set 'Year' as the index
df_ind = df[df['Country'] == 'India'].set_index('Year')

# Display the filtered DataFrame with 'Year' as the index
df_ind
```

```python
# Drop non-numeric columns like 'Country'
df_ind_numeric = df_ind.drop(columns=['Country'])
df_ind_numeric
```

```python
# Assume df_ind_numeric is your numeric DataFrame
median_values = df_ind_numeric.median()
mean_values = df_ind_numeric.mean()
variance_values = df_ind_numeric.var()
standard_deviation_values = df_ind_numeric.std()

# Create a DataFrame for the summary statistics
summary_stats = pd.DataFrame({
    'Median': median_values,
    'Mean': mean_values,
    'Variance': variance_values,
    'Standard Deviation': standard_deviation_values
})

# Display the summary statistics
print(summary_stats)
```
